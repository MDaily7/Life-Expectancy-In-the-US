```python
# create totals dfs
health_totals_df = health_measures_df[health_total_cols].copy()
additional_totals_df = additional_measures_df[additional_total_cols].copy()
```

```python
# Create state total dataframes
health_state_df = health_totals_df.loc[health_totals_df[('County', 'County')].isnull()]
additional_state_df = additional_totals_df.loc[additional_totals_df[('County', 'County')].isnull()]
```

```python
# remove county column from state dataframes and state rows from totals dfs
health_totals_df = health_totals_df.loc[health_totals_df[('County', 'County')].notnull()]
additional_totals_df = additional_totals_df.loc[additional_totals_df[('County', 'County')].notnull()]
health_state_df.drop(('County', 'County'), inplace=True, axis=1)
additional_state_df.drop(('County', 'County'), inplace=True, axis=1)
```

```python
# Find nulls in health totals df
print(health_totals_df.info(verbose=True, show_counts=True))
```

```python
# Fill in nulls with the mean of each numeric column for health totals df
num_col = [('Premature death', 'Deaths'), ('Low birthweight', '% Low birthweight'),('Food environment index', 'Food Environment Index'),('Access to exercise opportunities', '% With Access to Exercise Opportunities'),('Alcohol-impaired driving deaths', '# Alcohol-Impaired Driving Deaths'),('Sexually transmitted infections', '# Chlamydia Cases'),('Teen births', 'Teen Birth Rate'),('Uninsured', '# Uninsured'),('Primary care physicians', '# Primary Care Physicians'),('Dentists', '# Dentists'),('Mental health providers', '# Mental Health Providers'),('Preventable hospital stays', 'Preventable Hospitalization Rate'),('Mammography screening', '% With Annual Mammogram'),('Flu vaccinations', '% Vaccinated'),('Unemployment', '# Unemployed'),('Children in poverty', '% Children in Poverty'),('Income inequality', '80th Percentile Income'),('Children in single-parent households','# Children in Single-Parent Households'),('Violent crime', 'Annual Average Violent Crimes'),('Injury deaths', '# Injury Deaths'),('Air pollution - particulate matter', 'Average Daily PM2.5'),('Driving alone to work', '% Drive Alone to Work')]
for col in num_col:
    health_totals_df[col]=pd.to_numeric(health_totals_df[col])
    health_totals_df[col].fillna(health_totals_df[col].mean(), inplace=True)
health_totals_df.head()
```
